```python
# Question 3
# Which Male and Female Professor has the highest and the lowest salaries?
highest_male_salary = data[data['sex'] == 'Male']['salary'].max()
lowest_male_salary = data[data['sex'] == 'Male']['salary'].min()
highest_female_salary = data[data['sex'] == 'Female']['salary'].max()
lowest_female_salary = data[data['sex'] == 'Female']['salary'].min()

print("\nMale Professor with the highest salary:", highest_male_salary)
print("Male Professor with the lowest salary:", lowest_male_salary)
print("Female Professor with the highest salary:", highest_female_salary)
print("Female Professor with the lowest salary:", lowest_female_salary)
```

```python
# Question 4
# Which Professor takes the highest and lowest salaries?
highest_salary_professor = data.loc[data['salary'].idxmax()]
lowest_salary_professor = data.loc[data['salary'].idxmin()]

print("\nProfessor with the highest salary:")
print(highest_salary_professor)
print("\nProfessor with the lowest salary:")
print(lowest_salary_professor)
```

```python
# Question 5
# Missing Salaries - should be replaced as mean of the matching salaries of those whose service is the same
data['salary'] = data.groupby('service')['salary'].transform(lambda x: x.fillna(x.mean()))
```

```python
# Question 6
# Missing phd - should be replaced as mean of the matching service
data['phd'] = data.groupby('service')['phd'].transform(lambda x: x.fillna(x.mean()))
```

```python
# Rechecking if there are still missing values
missing_values = data.isnull().sum()
print("\nMissing values after execution")
print(missing_values)
```
