1. Read Total profit of all months and show it using a line plot

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("company_sales_data.csv")
plt.plot(data["month_number"],data["total_profit"])
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Total Profit per Month")
plt.grid(True)
plt.show()
```

2. Get total profit of all months and show line plot with the following Style properties.
Generated line plot must include following Style properties:
- Line Style dotted and Line-color should be red.
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

```python
plt.plot(data["month_number"],data["total_units"],label="Sold Units",linestyle="dotted",color="red",marker="o",markerfacecolor="red",markersize=5,linewidth=3)
plt.xlabel("Month Number")
plt.ylabel("Sold Units")
plt.title("Total Sold Units per Month")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

3. Read all product sales data and show it using a multiline plot. Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
for product in data.columns[2:]:
    sales_data = data[product]
    plt.plot(data["month_number"], sales_data, label=product)
plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.title("Monthly Sales per Product")
plt.legend()
plt.grid(True)
plt.show()
```

4. Read face cream and facewash product sales data and show it using the bar chart.

```python
plt.bar(["Face Cream", "Facewash"], [data["facecream"].sum(),data["facewash"].sum()])
plt.xlabel("Products")
plt.ylabel("Total Units Sold")
plt.title("Total Sales of Face Cream and Facewash")
plt.show()
```

5. Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk.

```python
plt.bar(data["month_number"],data["bathingsoap"])
plt.show()
```

6. Read the total profit of each month and show it using the histogram to see the most common profit
ranges.

```python
plt.hist(data["total_profit"],bins=10,edgecolor='black')
plt.xlabel("Total Profit")
plt.ylabel("Frequency")
plt.title("Distribution of Total Profit per Month")
plt.grid(True)
plt.show()
```

7. Calculate total sale data for last year for each product and show it using a Pie chart.

```python
plt.pie(data.iloc[:,1:].sum(),labels=data.columns[1:],autopct="%1.1f%%",shadow=True)
plt.title("Total Sales per Product")
plt.show()
```

8. Read Bathing soap facewash of all months and display it using the Subplot. 

```python
month_numbers = data["month_number"]
bathing_soap_sales = data["bathingsoap"]
facewash_sales = data["facewash"]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))  # Adjust figsize for readability
ax1.plot(month_numbers, bathing_soap_sales)
ax1.set_xlabel("Month Number")
ax1.set_ylabel("Bathing Soap Units Sold")
ax1.set_title("Bathing Soap Sales")
ax1.grid(True)
ax2.plot(month_numbers, facewash_sales)
ax2.set_xlabel("Month Number")
ax2.set_ylabel("Facewash Units Sold")
ax2.set_title("Facewash Sales")
ax2.grid(True)
plt.tight_layout()
plt.show()
```

9. Read all product sales data and show it using the stack plot.

```python
plt.stackplot(data["month_number"],data.iloc[:,1:].transpose(), labels=data.columns[1:])
plt.xlabel("Month Number")
plt.ylabel("Total Units Sold")
plt.title("Monthly Sales Breakdown by Product")
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
