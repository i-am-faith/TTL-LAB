1. Write a Python program to create a pie chart of the popularity of programming Languages.
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
i. Pull the &quot;Python&quot; wedge 0.3 from the center of the pie
ii. Add a legend
iii.Add a legend with a header
iv. Specify a new color for each wedge
v. Add a shadow to the pie chart
vi. Start the first wedge at 90 degrees

```python
import matplotlib.pyplot as plt
plt.pie([22.2,17.6,8.8,8,7.7,6.7],explode=[0,0.3,0,0,0,0],labels=['Java','Python','PHP','JavaScript','C#','C++']
,autopct="%1.1f%%",startangle=90,colors=['blue','green','red','purple','gold','cyan'],shadow=True)
plt.legend(title="Programming Language Popularity")
plt.axis('equal')
plt.show()
```

4. Write a Python programming to display a bar chart of the popularity of programming Languages. Use
different color for each bar. Attach a text label above each bar displaying its popularity (float value).
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
plt.figure(figsize=(8, 6))
plt.bar(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],[22.2, 17.6, 8.8, 8, 7.7, 6.7], color=['blue', 'green', 'red', 'purple', 'gold', 'cyan'])
for i, v in enumerate([22.2, 17.6, 8.8, 8, 7.7, 6.7]):
    plt.text(i, v + 0.1, f"{v:.1f}", ha='center', va='bottom', fontsize=12)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Programming Languages')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
